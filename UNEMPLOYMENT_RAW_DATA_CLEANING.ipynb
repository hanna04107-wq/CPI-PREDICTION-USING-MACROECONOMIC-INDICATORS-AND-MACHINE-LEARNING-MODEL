```python
import pandas as pd
import numpy as np
```

```python
# 1. Load the raw unemployment data
file_path_unemp = "/content/Unemployment-dataset - Unemployment.csv"
df_unemp = pd.read_csv(file_path_unemp, usecols=['date', 'u_rate'])
```

```python
# 2. View the first 5 rows of unemployment data
df_unemp.head()
```

```python
# 3. Fix the date format dynamically using the Maximum trick
parsed_dates = pd.to_datetime(df_unemp['date'], format='mixed', dayfirst=True)
df_unemp['Year'] = parsed_dates.dt.year.astype(int)
```

```python
# 4. Extract the true month by taking the maximum of the parsed month
df_unemp['Month'] = np.maximum(parsed_dates.dt.month, parsed_dates.dt.day).astype(int)
```

```python
# 5. Create our standard Master 'Date' column (YYYY-MM-01)
df_unemp['Date'] = pd.to_datetime(df_unemp[['Year', 'Month']].assign(DAY=1))
```

```python
# 6. View the first 5 rows of unemployment data after creating date column
df_unemp.head()
```

```python
# 6. Filter strictly for the 2013 to 2026 Jan timeframe
df_unemp_clean = df_unemp[(df_unemp['Date'] >= '2013-01-01') & (df_unemp['Date'] <= '2026-01-01')].copy()
```

```python
# 7. Keep only the essential columns
df_unemp_clean = df_unemp_clean[['Date', 'Year', 'Month', 'u_rate']].reset_index(drop=True)
```

```python
# 8. View the first 5 rows of unemployment data after selecting necessary columns only
df_unemp_clean.head()
```

```python
# 9. Rename 'u_rate' to proper name
df_unemp_clean.rename(columns={'u_rate': 'Unemployment_Rate'}, inplace=True)
```

```python
print("Unemployment Data (First 5 Rows):")
print(df_unemp_clean.head())
print("\nUnemployment Data (Last 5 Rows):")
print(df_unemp_clean.tail())
print(f"\nTotal Rows: {len(df_unemp_clean)}")
```

```python
df_unemp_clean = df_unemp_clean.dropna()
print("DataFrame after dropping rows with null values:")
print(df_unemp_clean.head())
print(f"Total Rows after dropping nulls: {len(df_unemp_clean)}")
```
