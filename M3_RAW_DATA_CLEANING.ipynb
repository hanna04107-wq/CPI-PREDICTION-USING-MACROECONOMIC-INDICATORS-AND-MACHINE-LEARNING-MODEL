```python
import pandas as pd
import numpy as np
```

```python
# 1. Load the raw M3 data
file_path_m3 = "/content/M3-dataset - M3.csv"
```

```python
# 2. We skip the first 6 messy header rows and only take Column 0 (Year), 1 (Month), and 2 (Total M3)
df_m3 = pd.read_csv(file_path_m3, usecols=[0, 1, 2], skiprows=6, header=None)
df_m3.columns = ['Year', 'Month', 'M3_Total']
```

```python
# 3. View the first 5 rows of M3 data
df_m3.head()
```

```python
# 4. Clean the Year column and forward-fill the empty blanks
df_m3['Year'] = pd.to_numeric(df_m3['Year'], errors='coerce').ffill()
```

```python
# 5. View the first 5 rows of M3 data
df_m3.head()
```

```python
# 6. Clean the Month column and drop any purely text/empty rows
df_m3['Month'] = pd.to_numeric(df_m3['Month'], errors='coerce')
df_m3 = df_m3.dropna(subset=['Month'])
```

```python
# 7. Clean the M3 values (in case any numbers were read as strings with commas)
if df_m3['M3_Total'].dtype == 'O':
    df_m3['M3_Total'] = df_m3['M3_Total'].str.replace(',', '')
df_m3['M3_Total'] = pd.to_numeric(df_m3['M3_Total'], errors='coerce')
```

```python
# 8. Convert Year and Month to clean integers
df_m3['Year'] = df_m3['Year'].astype(int)
df_m3['Month'] = df_m3['Month'].astype(int)
```

```python
# 9. View the first 5 rows of M3 data after column's data type change
df_m3.head()
```

```python
# 10. Create our 'Date' column for merging
df_m3['Date'] = pd.to_datetime(df_m3[['Year', 'Month']].assign(DAY=1))
```

```python
# 11. Filter strictly for 2013 to 2026 timeframe
df_m3_clean = df_m3[(df_m3['Date'] >= '2013-01-01') & (df_m3['Date'] <= '2026-01-01')].copy()
```

```python
# 12. Keep only the essential columns to merge later
df_m3_clean = df_m3_clean[['Date', 'Year', 'Month', 'M3_Total']].reset_index(drop=True)
```

```python
print("M3 Data (First 5 Rows):")
print(df_m3_clean.head())
print("\nM3 Data (Last 5 Rows):")
print(df_m3_clean.tail())
print(f"\nTotal Rows: {len(df_m3_clean)}")
```
