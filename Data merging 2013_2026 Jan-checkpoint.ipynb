```python
import pandas as pd
import numpy as np

# 1. Load the raw CPI data
file_path_cpi = "C:/Users/User/Downloads/CPI-dataset - CPI.csv"
df_cpi = pd.read_csv(file_path_cpi, usecols=['date', 'division', 'index'])

# 2. Filter ONLY for the 'overall' division right from the start
df_cpi = df_cpi[df_cpi['division'] == 'overall'].copy()

# 3. Fix the date format dynamically using the Maximum trick
parsed_dates = pd.to_datetime(df_cpi['date'], format='mixed', dayfirst=True)
df_cpi['Year'] = parsed_dates.dt.year.astype(int)

# Extract the true month by taking the max of the month or day
df_cpi['Month'] = np.maximum(parsed_dates.dt.month, parsed_dates.dt.day).astype(int)

# 4. Create our standard Master 'Date' column
df_cpi['Date'] = pd.to_datetime(df_cpi[['Year', 'Month']].assign(DAY=1))

# 5. Rename the original 'index' column to 'CPI_Index' to avoid confusion
df_cpi.rename(columns={'index': 'CPI_Index'}, inplace=True)

# 6. Keep only the necessary columns
df_cpi_clean = df_cpi[['Date', 'Year', 'Month', 'CPI_Index']].copy()

# 7. Filter strictly for the 2013 to 2026 timeframe and sort chronologically
df_cpi_clean = df_cpi_clean[(df_cpi_clean['Date'] >= '2013-01-01') & (df_cpi_clean['Date'] <= '2026-01-01')].reset_index(drop=True)
df_cpi_clean = df_cpi_clean.sort_values(by='Date').reset_index(drop=True)

print("CPI Data (First 5 Rows):")
print(df_cpi_clean.head())
print("\nCPI Data (Last 5 Rows):")
print(df_cpi_clean.tail())
print(f"\nTotal Rows: {len(df_cpi_clean)}")
```

```python
import pandas as pd

# 1. Load the raw OPR data
file_path_opr = "C:/Users/User/Downloads/OPR-dataset - OPR.csv"
df_opr = pd.read_csv(file_path_opr, header=0, usecols=['Date', 'Rate'])

# 2. Clean the raw Dates and drop empty rows
df_opr['Date'] = pd.to_datetime(df_opr['Date'], errors='coerce')
df_opr = df_opr.dropna(subset=['Date', 'Rate']).sort_values('Date')

# 3. Extract Year and Month to easily match our master timeline
df_opr['Year'] = df_opr['Date'].dt.year
df_opr['Month'] = df_opr['Date'].dt.month

# If BNM somehow changes the rate twice in one month, we just keep the newest one
df_opr_unique = df_opr.drop_duplicates(subset=['Year', 'Month'], keep='last')[['Year', 'Month', 'Rate']]

# 4. Create a Master Timeline (Starting from 2004-05-01 to capture the oldest rate baseline)
timeline = pd.DataFrame({'Date': pd.date_range(start='2004-05-01', end='2026-02-01', freq='MS')})
timeline['Year'] = timeline['Date'].dt.year
timeline['Month'] = timeline['Date'].dt.month

# 5. Merge the BNM OPR changes onto our perfect timeline
merged_opr = pd.merge(timeline, df_opr_unique, on=['Year', 'Month'], how='left')

# 6. Apply the Forward Fill to drag the rate down across empty months
merged_opr['OPR_Rate'] = merged_opr['Rate'].ffill()

# 7. Finally, filter it exactly for the 2013 to 2026 timeframe! (BUG FIXED HERE)
df_opr_clean = merged_opr[(merged_opr['Date'] >= '2013-01-01') & (merged_opr['Date'] <= '2026-01-01')].copy()

# Keep only the essential columns for merging later
df_opr_clean = df_opr_clean[['Date', 'Year', 'Month', 'OPR_Rate']].reset_index(drop=True)

print("OPR Data (First 5 Rows):")
print(df_opr_clean.head())
df_opr_clean
print("\nOPR Data (Last 5 Rows):")
print(df_opr_clean.tail())
print(f"\nTotal Rows: {len(df_opr_clean)}")
```

```python
import pandas as pd
import numpy as np

# 1. Load the raw Unemployment data
file_path_unemp = "C:/Users/User/Downloads/Unemployment-dataset - Unemployment.csv"
# Only read the two columns we care about
df_unemp = pd.read_csv(file_path_unemp, usecols=['date', 'u_rate'])

# 2. Fix the date format dynamically using the Maximum trick
# This prevents your local Excel settings from shuffling the months and days!
parsed_dates = pd.to_datetime(df_unemp['date'], format='mixed', dayfirst=True)
df_unemp['Year'] = parsed_dates.dt.year.astype(int)

# Extract the true month by taking the maximum of the parsed month or day
df_unemp['Month'] = np.maximum(parsed_dates.dt.month, parsed_dates.dt.day).astype(int)

# 3. Create our standard Master 'Date' column (YYYY-MM-01)
df_unemp['Date'] = pd.to_datetime(df_unemp[['Year', 'Month']].assign(DAY=1))

# 4. Filter strictly for the 2013 to 2024 timeframe
df_unemp_clean = df_unemp[(df_unemp['Date'] >= '2013-01-01') & (df_unemp['Date'] <= '2026-01-01')].copy()

# Keep only the essential columns to merge later
df_unemp_clean = df_unemp_clean[['Date', 'Year', 'Month', 'u_rate']].reset_index(drop=True)

# Rename 'u_rate' so it's super clear when we merge everything
df_unemp_clean.rename(columns={'u_rate': 'Unemployment_Rate'}, inplace=True)

print("Unemployment Data (First 5 Rows):")
print(df_unemp_clean.head())
print(f"Total Rows: {len(df_unemp_clean)}") # Should be perfectly 144!
```

```python
import pandas as pd

# 1. Load the raw M3 data 
file_path_m3 = "C:/Users/User/Downloads/M3-dataset - M3.csv"

# 2. We skip the first 5 messy header rows and only take Column 0 (Year), 1 (Month), and 2 (Total M3)
df_m3 = pd.read_csv(file_path_m3, usecols=[0, 1, 2], skiprows=5, header=None)
df_m3.columns = ['Year', 'Month', 'M3_Total']

# 3. Clean the Year column and forward-fill the empty blanks
df_m3['Year'] = pd.to_numeric(df_m3['Year'], errors='coerce').ffill()

# 4. Clean the Month column and drop any purely text/empty rows
df_m3['Month'] = pd.to_numeric(df_m3['Month'], errors='coerce')
df_m3 = df_m3.dropna(subset=['Month'])

# 5. Clean the M3 values (in case any numbers were read as strings with commas)
if df_m3['M3_Total'].dtype == 'O':
    df_m3['M3_Total'] = df_m3['M3_Total'].str.replace(',', '')
df_m3['M3_Total'] = pd.to_numeric(df_m3['M3_Total'], errors='coerce')

# Convert Year and Month to clean integers
df_m3['Year'] = df_m3['Year'].astype(int)
df_m3['Month'] = df_m3['Month'].astype(int)

# 6. Create our Master 'Date' column for merging
df_m3['Date'] = pd.to_datetime(df_m3[['Year', 'Month']].assign(DAY=1))

# 7. Filter strictly for our 2013 to 2024 timeframe
df_m3_clean = df_m3[(df_m3['Date'] >= '2013-01-01') & (df_m3['Date'] <= '2026-01-01')].copy()

# Keep only the essential columns to merge later
df_m3_clean = df_m3_clean[['Date', 'Year', 'Month', 'M3_Total']].reset_index(drop=True)

print("M3 Data (First 5 Rows):")
print(df_m3_clean.head())
print(f"Total Rows: {len(df_m3_clean)}") # Should be exactly 144!
```

```python
import pandas as pd
import numpy as np

# 1. Load the raw PPI data
file_path_ppi = "C:/Users/User/Downloads/PPI-dataset - PPI.csv"
df_ppi = pd.read_csv(file_path_ppi, usecols=['series', 'date', 'index'])

# 2. Fix the date format dynamically using the Maximum trick
parsed_dates = pd.to_datetime(df_ppi['date'], format='mixed', dayfirst=True)
df_ppi['Year'] = parsed_dates.dt.year.astype(int)

# Extract the true month by taking the max of the month or day
df_ppi['Month'] = np.maximum(parsed_dates.dt.month, parsed_dates.dt.day).astype(int)

# 3. Create our standard Master 'Date' column
df_ppi['Date'] = pd.to_datetime(df_ppi[['Year', 'Month']].assign(DAY=1))

# 4. Pivot the stacked data into separate side-by-side columns
df_ppi_pivot = df_ppi.pivot(index=['Date', 'Year', 'Month'], columns='series', values='index').reset_index()

# 5. Keep only the index and year-growth columns, and rename them clearly
df_ppi_clean = df_ppi_pivot[['Date', 'Year', 'Month', 'abs']].copy()
df_ppi_clean.rename(columns={'abs': 'PPI_Index'}, inplace=True)

# 6. Filter strictly for the 2013 to 2024 timeframe
df_ppi_clean = df_ppi_clean[(df_ppi_clean['Date'] >= '2013-01-01') & (df_ppi_clean['Date'] <= '2026-01-01')].reset_index(drop=True)

# Remove the 'series' column label created by the pivot
df_ppi_clean.columns.name = None

print("PPI Data (First 5 Rows):")
print(df_ppi_clean.head())
print(f"Total Rows: {len(df_ppi_clean)}")
```

```python
import pandas as pd

# 1. Load the raw Exchange Rate data (without headers so we can count columns easily)
file_path_ex = "C:/Users/User/Downloads/Exchange-dataset - Exchange rate.csv"
df_ex = pd.read_csv(file_path_ex, header=None)

# 2. Extract Column 1 (Year), Column 2 (Month), and Column 16 (USD Average)
df_ex_clean = df_ex.iloc[:, [0, 1, 15]].copy()
df_ex_clean.columns = ['Year', 'Month', 'USD_Avg']

# 3. Convert Year and Month to numbers (this turns any text headers into NaN)
df_ex_clean['Year'] = pd.to_numeric(df_ex_clean['Year'], errors='coerce')
df_ex_clean['Month'] = pd.to_numeric(df_ex_clean['Month'], errors='coerce')

# 4. Drop rows where the Month is NaN (This perfectly removes the annual averages at the top of the file)
df_ex_clean = df_ex_clean.dropna(subset=['Month'])

# 5. Forward-fill the Year column down across the empty months
df_ex_clean['Year'] = df_ex_clean['Year'].ffill()

# 6. Convert to clean numbers
df_ex_clean['Year'] = df_ex_clean['Year'].astype(int)
df_ex_clean['Month'] = df_ex_clean['Month'].astype(int)
df_ex_clean['USD_Avg'] = pd.to_numeric(df_ex_clean['USD_Avg'], errors='coerce')

# 7. Drop footnote rows at the very bottom of the file that lack actual exchange rates
df_ex_clean = df_ex_clean.dropna(subset=['USD_Avg'])

# 8. Filter for 2013 to 2024 and create our Master Date
df_ex_clean['Date'] = pd.to_datetime(df_ex_clean[['Year', 'Month']].assign(DAY=1))
df_ex_clean = df_ex_clean[(df_ex_clean['Date'] >= '2013-01-01') & (df_ex_clean['Date'] <= '2026-01-01')].reset_index(drop=True)

# Keep essential columns and rename Date for merging
df_ex_clean = df_ex_clean[['Date', 'Year', 'Month', 'USD_Avg']]

print("Exchange Rate Data (First 5 Rows):")
print(df_ex_clean.head())
print(f"Total Rows: {len(df_ex_clean)}") # Should be perfectly 144!
```

```python
import pandas as pd
from functools import reduce

# 1. Put all of your perfectly cleaned dataframes into a single list
# (This assumes you have run all the previous blocks in your notebook/environment)
dataframes = [
    df_cpi_clean, 
    df_opr_clean, 
    df_unemp_clean, 
    df_m3_clean, 
    df_ppi_clean, 
    df_ex_clean
]

# 2. Merge them all together!
# We use an 'inner' merge on our exact time identifiers to snap them perfectly side-by-side
df_merge = reduce(lambda left, right: pd.merge(left, right, on=['Date', 'Year', 'Month'], how='inner'), dataframes)

# 3. Double-check for any accidental missing values before we save it
print("Checking for missing values in the Master Dataset:")
print(df_merge.isnull().sum()) # These should all be 0!

print("\nMerged Dataset Shape:", df_merge.shape) # Should be (144, 9)

# 4. View the final product
print("\nFinal Machine Learning Dataset (First 5 Rows):")
print(df_merge.head())

# 5. Export this to a final CSV so you never have to clean the raw data again!
df_merge.to_csv("Final_Dataset_2013_2026_Jan.csv", index=False)
print("\nSuccess! Saved to 'Final_CPI_Dataset_2013_2026_Jan.csv'")
```
