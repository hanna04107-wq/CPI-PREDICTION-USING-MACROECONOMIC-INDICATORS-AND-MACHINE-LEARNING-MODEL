```python
import pandas as pd
```

```python
# 1. Load all your cleaned datasets individually
df_cpi   = pd.read_csv("Clean_CPI.csv")
df_ex    = pd.read_csv("Clean_Exchange_Rate.csv")
df_m3    = pd.read_csv("Clean_M3.csv")
df_opr   = pd.read_csv("Clean_OPR.csv")
df_ppi   = pd.read_csv("Clean_PPI.csv")
df_unemp = pd.read_csv("Clean_Unemployment.csv")
```

```python
# 2. Merge the dataset
master_df = df_cpi.merge(df_ex, on=['Date', 'Year', 'Month'], how='inner')
master_df = master_df.merge(df_m3, on=['Date', 'Year', 'Month'], how='inner')
master_df = master_df.merge(df_opr, on=['Date', 'Year', 'Month'], how='inner')
master_df = master_df.merge(df_ppi, on=['Date', 'Year', 'Month'], how='inner')
master_df = master_df.merge(df_unemp, on=['Date', 'Year', 'Month'], how='inner')
```

```python
# 4. Save finalized dataset
master_df.to_csv("Final_Macroeconomic_Dataset_2013_2026_Jan.csv", index=False)
```
