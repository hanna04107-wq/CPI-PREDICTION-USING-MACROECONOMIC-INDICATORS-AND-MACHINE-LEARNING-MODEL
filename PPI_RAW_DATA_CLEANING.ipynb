```python
import pandas as pd
import numpy as np
```

```python
# 1. Load the raw PPI data
file_path_ppi = "/content/PPI-dataset - PPI.csv"
df_ppi = pd.read_csv(file_path_ppi, usecols=['series', 'date', 'index'])
```

```python
# 2. View the first 5 rows of PPI data
df_ppi.head()
```

```python
# 3. Fix the date format dynamically using the Maximum trick
parsed_dates = pd.to_datetime(df_ppi['date'], format='mixed', dayfirst=True)
df_ppi['Year'] = parsed_dates.dt.year.astype(int)
```

```python
# 4. Extract the month by taking the max of the month or day
df_ppi['Month'] = np.maximum(parsed_dates.dt.month, parsed_dates.dt.day).astype(int)
```

```python
# 5. Create standard Master 'Date' column
df_ppi['Date'] = pd.to_datetime(df_ppi[['Year', 'Month']].assign(DAY=1))
```

```python
# 6. View the first and last 5 rows of PPI data after inserting Date column
print(df_ppi.head())
print(df_ppi.tail())
```

```python
# 7. Pivot the stacked data into separate side-by-side columns
df_ppi_pivot = df_ppi.pivot(index=['Date', 'Year', 'Month'], columns='series', values='index').reset_index()
```

```python
# 8. Keep only the index and rename them clearly
df_ppi_clean = df_ppi_pivot[['Date', 'Year', 'Month', 'abs']].copy()
df_ppi_clean.rename(columns={'abs': 'PPI_Index'}, inplace=True)
```

```python
# 9. Filter strictly for the 2013 to 2026 timeframe
df_ppi_clean = df_ppi_clean[(df_ppi_clean['Date'] >= '2013-01-01') & (df_ppi_clean['Date'] <= '2026-01-01')].reset_index(drop=True)
```

```python
# 10. View the first and last 5 rows of PPI data after setting the timeframe
print(df_ppi_clean.head())
print(df_ppi_clean.tail())
```

```python
# 11.Remove the 'series' column label created by the pivot
df_ppi_clean.columns.name = None
```

```python
print("PPI Data (First 5 Rows):")
print(df_ppi_clean.head())
print("\nPPI Data (Last 5 Rows):")
print(df_ppi_clean.tail())
print(f"\nTotal Rows: {len(df_ppi_clean)}")
```

```python
df_ppi_clean = df_ppi_clean.dropna()
print("DataFrame after dropping rows with null values:")
print(df_ppi_clean.head())
print(f"Total Rows after dropping nulls: {len(df_ppi_clean)}")
```
