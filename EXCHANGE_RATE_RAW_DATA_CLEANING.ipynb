```python
import pandas as pd
import numpy as np
```

```python
# 1. Load the raw Exchange Rate data (without headers so we can count columns easily)
file_path_ex = "/content/Exchange-dataset - Exchange rate.csv"
df_ex = pd.read_csv(file_path_ex, header=None)
```

```python
# 2. Extract Column 1 (Year), Column 2 (Month), and Column 16 (USD Average)
df_ex_clean = df_ex.iloc[:, [0, 1, 15]].copy()
df_ex_clean.columns = ['Year', 'Month', 'USD_Avg']
```

```python
# 3. View the first 5 rows of Exchange data
df_ex_clean.head()
```

```python
# 4. Convert Year and Month to numbers (this turns any text headers into NaN)
df_ex_clean['Year'] = pd.to_numeric(df_ex_clean['Year'], errors='coerce')
df_ex_clean['Month'] = pd.to_numeric(df_ex_clean['Month'], errors='coerce')
```

```python
# 5. View the first 5 rows of Exchange data
df_ex_clean.head()
```

```python
# 6. Drop rows where the Month is NaN (This perfectly removes the annual averages at the top of the file)
df_ex_clean = df_ex_clean.dropna(subset=['Month'])
```

```python
# 7. View the first 5 rows of Exchange data after dropping rows where the Month is NAN
df_ex_clean.head()
```

```python
# 8. Forward-fill the Year column down across the empty months
df_ex_clean['Year'] = df_ex_clean['Year'].ffill()
```

```python
# 9. View the first 5 rows of Exchange data after forward fill of year
df_ex_clean.head()
```

```python
# 10. Convert to clean numbers
df_ex_clean['Year'] = df_ex_clean['Year'].astype(int)
df_ex_clean['Month'] = df_ex_clean['Month'].astype(int)
df_ex_clean['USD_Avg'] = pd.to_numeric(df_ex_clean['USD_Avg'], errors='coerce')
```

```python
# 11. View the first and last 5 rows of Exchange data
print(df_ex_clean.head())
print(df_ex_clean.tail())
```

```python
# 12. Drop footnote rows at the very bottom of the file that lack actual exchange rates
df_ex_clean = df_ex_clean.dropna(subset=['USD_Avg'])
```

```python
# 13. View the first and last 5 rows of Exchange data after dropping footnote
print(df_ex_clean.head())
print(df_ex_clean.tail())
```

```python
# 14. Create standard Master 'Date' column
df_ex_clean['Date'] = pd.to_datetime(df_ex_clean[['Year', 'Month']].assign(DAY=1))
```

```python
# 15. Filter strictly for the 2013 to 2026 timeframe
df_ex_clean = df_ex_clean[(df_ex_clean['Date'] >= '2013-01-01') & (df_ex_clean['Date'] <= '2026-01-01')].reset_index(drop=True)
```

```python
# 16. View the first and last 5 rows of Exchange data after setting timeframe
print(df_ex_clean.head())
print(df_ex_clean.tail())
```

```python
# 17. Keep essential columns and rename Date for merging
df_ex_clean = df_ex_clean[['Date', 'Year', 'Month', 'USD_Avg']]
```

```python
print("Exchange Data (First 5 Rows):")
print(df_ex_clean.head())
print("\nExchange Data (Last 5 Rows):")
print(df_ex_clean.tail())
print(f"\nTotal Rows: {len(df_ex_clean)}")
```

```python
df_ex_clean = df_ex_clean.dropna()
print("DataFrame after dropping rows with null values:")
print(df_ex_clean.head())
print(f"Total Rows after dropping nulls: {len(df_ex_clean)}")
```
