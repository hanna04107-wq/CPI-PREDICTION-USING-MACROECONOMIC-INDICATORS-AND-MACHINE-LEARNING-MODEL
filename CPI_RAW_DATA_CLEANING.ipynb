```python
import pandas as pd
import numpy as np
```

```python
# 1. Load the raw CPI data
file_path_cpi = "/content/CPI-dataset - CPI.csv"
df_cpi = pd.read_csv(file_path_cpi, usecols=['date', 'division', 'index'])
```

```python
# 2. View the first 5 rows of CPI data
df_cpi.head()
```

```python
# 3. Filter only for the 'overall' division right from the start
df_cpi = df_cpi[df_cpi['division'] == 'overall'].copy()
```

```python
# 4. Fix the date format dynamically using the Maximum trick
parsed_dates = pd.to_datetime(df_cpi['date'], format='mixed', dayfirst=True)
df_cpi['Year'] = parsed_dates.dt.year.astype(int)
```

```python
# 5. Extract the true month by taking the max of the month
df_cpi['Month'] = np.maximum(parsed_dates.dt.month, parsed_dates.dt.day).astype(int)
```

```python
# 6. Create our standard Master 'Date' column
df_cpi['Date'] = pd.to_datetime(df_cpi[['Year', 'Month']].assign(DAY=1))
```

```python
# 7. Rename the original 'index' column to 'CPI_Index' to avoid confusion
df_cpi.rename(columns={'index': 'CPI_Index'}, inplace=True)
```

```python
# 8. View the first 5 rows after fixing the date format , extracting the month , craeting date column
df_cpi.head()
```

```python
# 9. Keep only the necessary columns
df_cpi_clean = df_cpi[['Date', 'Year', 'Month', 'CPI_Index']].copy()
```

```python
# 10. View the first 5 rows after selecting necessary columns only
df_cpi_clean.head()
```

```python
# 11. Filter strictly for the 2013 to 2026 timeframe and sort chronologically
df_cpi_clean = df_cpi_clean[(df_cpi_clean['Date'] >= '2013-01-01') & (df_cpi_clean['Date'] <= '2026-01-01')].reset_index(drop=True)
df_cpi_clean = df_cpi_clean.sort_values(by='Date').reset_index(drop=True)
```

```python
print("CPI Data (First 5 Rows):")
print(df_cpi_clean.head())
print("\nCPI Data (Last 5 Rows):")
print(df_cpi_clean.tail())
print(f"\nTotal Rows: {len(df_cpi_clean)}")
```
